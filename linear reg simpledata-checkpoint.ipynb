```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data = data.drop(298)
data = data.drop(299)
```

```python
X, Y = data["X"],data["Y"]
```

```python
plt.scatter(X,Y, marker='x', color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y)
X_train = X_train[:, np.newaxis]
X_test = y_test[:, np.newaxis]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test, y_test)*100     # testing score
```

```python
model.score(X_train, y_train)*100   # training score
```

```python
# comparing y_test with predicted y
c = [i for i in range(len(y_test))]
plt.plot(c,y_test, color='red')
plt.plot(c, y_pred)
```

```python
#plotting the regression line
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred)
```
